# Easy example to learn some basics

# Import the packages

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

# Define function to fit

```python
def myfunc(x):
 return slope * x + intercept
```

# Import data

```python
full_health_data = pd.read_csv("data.csv", header=0, sep=",")
```

```python
print(full_health_data.info())
print(full_health_data.head())
```

```python
print(full_health_data.describe()) 
```

# Analyze the data

```python
# Data
x = full_health_data["Average_Pulse"]
y = full_health_data["Calorie_Burnage"]

# Linear regression of the data
slope, intercept, r, p, std_err = stats.linregress(x, y)

# Calculate the fitted curve with the data obtained
mymodel = list(map(myfunc, x))
```

# Plot the result
Example of bad fitting

```python
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.ylim(ymin=0, ymax=2000)
plt.xlim(xmin=0, xmax=200)
plt.xlabel("Average_Pulse")
plt.ylabel ("Calorie_Burnage")
plt.show()

print('R-square value =',r)
print('P-value =',p)
```

# Example of good fitting

```python
# Data
x = full_health_data["Duration"]
y = full_health_data["Calorie_Burnage"]

# Linear regression of the data
slope, intercept, r, p, std_err = stats.linregress(x, y)

# Calculate the fitted curve with the data obtained
mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.ylim(ymin=0, ymax=2000)
plt.xlim(xmin=0, xmax=200)
plt.xlabel("Duration")
plt.ylabel ("Calorie_Burnage")
plt.show()

print('R-square value =',r)
print('P-value =',p)
```
